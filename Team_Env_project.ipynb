```python
# 폴더 안에 있는 xlsx 확장자를 가진 모든 파일 이름 리스트에 담기
import os
import pandas as pd
from glob import glob

path = r"C:\Users\lab\Desktop\AP_teamproject"
os.chdir(path)
data = glob("*.xlsx")
data
```

```python
# 사용할 수질 항목만 list에 저장하여 뽑아내기.
# 각 데이터에서 결측치 유무 확인 및 제거
# 중랑천4에서 각 항목마다 결측치 하나씩 발생

using_list = ['측정소명','년/월','수온(℃)','용존산소(㎎/L)','BOD(㎎/L)','수소이온농도','분원성대장균군수']
total_df = pd.DataFrame()
for i in range(len(data)):
    df = pd.read_excel(data[i])
    df = df[using_list]
    total_df = pd.concat([total_df,df],ignore_index=True)
    print(data[i], df.isnull().sum())

total_df = total_df.dropna()
```

```python
#용존산소 / 용존산소 포화농도 구하기 (WQI 산정할 때 DO index로 사용됨)
total_df["DO(% saturation)"] = (total_df['용존산소(㎎/L)'] / total_df["Do(100 %saturation)"])*100
```

```python
#max가 140이 넘으면 Equation에 적용할 수 없다.
total_df["DO(% saturation)"].describe()
```

```python
#DO(% Saturation) >= 140인 값에 대한 Equation은 없으므로 140인 값에 대한 index를 추출하고 해당 행을 삭제한다.

drop_index = total_df[total_df["DO(% saturation)"] >= 140].index
drop_index
total_df = total_df.drop(drop_index).reset_index().drop(['index'],axis=1)
total_df
```

```python
# DO index 구하기
x=[]
for row in total_df["DO(% saturation)"]:
    if 0 <= row < 40 :
        x1 = row*0.66 + 0.18
        x.append(x1)
    elif 40 <= row <100 :
        x1 = row*1.17 - 13.55
        x.append(x1)
    elif 100 <= row <140:
        x1 = 163.34 - row*0.62 
        x.append(x1)
total_df["Do_index"] = x
```

```python
# 분원성대장균 군수 (Fecal coli) index 구하기
import math
y=[]
for row in total_df["분원성대장균군수"]:
    if 1 <= row < 100 :
        y1 = -26.6*math.log(row) + 97.2
        y.append(y1)
    elif 100 <= row <10000 :
        y1 = -7.75*math.log(row) + 42.33
        y.append(y1)
    elif 10000 <= row:
        y1 = 2
        y.append(y1)
total_df["FC_index"] = y
```

```python
# pH index 구하기
w=[]
for row in total_df["수소이온농도"]:
    if 0.2 <= row < 0.5 :
        w1 = 16.1 + 7.35*row
        w.append(w1)
    elif 0.5 <= row <7.3:
        w1 = -142.367+33.5*row
        w.append(w1)
    elif 7.3 <= row < 10:
        w1 = 316.96-29.85*row
        w.append(w1)
    elif 10 <= row < 12 :
        w1 = 96.17-8*row
        w.append(w1)
    else : 
        w1 = 0
        w.append(w1)    
total_df["pH_index"] = w
```

```python
# BOD index 구하기
z=[]
for row in total_df["BOD(㎎/L)"]:
    if 0 <= row < 10 :
        z1 = 96.67-7*row
        z.append(z1)
    elif 10 <= row <30:
        z1 = 38.9-1.23*row
        z.append(z1)
    elif 30 <= row:
        z1 = 2
        z.append(z1)  
total_df["BOD_index"] = z
```

```python
# 위에서 구한 index로 WQI 산정
total_df["WQI(수질평가지수)"] = (total_df["Do_index"] * 0.31 + total_df["pH_index"] * 0.28 
                           + total_df["BOD_index"] * 0.22 + total_df["FC_index"] * 0.19)
total_df
```

```python
# WQI 지수에 따른 수질등급 나누기
class_WQI=[]

for row in total_df["WQI(수질평가지수)"]:
    if 63 < row <= 100 :
        class_WQI.append("A")
    elif 50 < row <= 63:
        class_WQI.append("B")
    elif 38 < row <= 50:
        class_WQI.append("C")
    elif row <= 38:
        class_WQI.append("D")
total_df["class_WQI"] = class_WQI
```

```python
from sklearn.linear_model import LinearRegression 
from sklearn.model_selection import train_test_split

X1 = df1[['수온(℃)', '용존산소(㎎/L)', 'BOD(㎎/L)','수소이온농도','분원성대장균군수']]
y1 = df1["WQI(수질평가지수)"]

X2 = df2[['수온(℃)', '용존산소(㎎/L)', 'BOD(㎎/L)','수소이온농도','분원성대장균군수']]
y2 = df2["WQI(수질평가지수)"]

X3 = df3[['수온(℃)', '용존산소(㎎/L)', 'BOD(㎎/L)','수소이온농도','분원성대장균군수']]
y3 = df3["WQI(수질평가지수)"]
```

```python
X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size = 0.2, random_state=0)
X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size = 0.2, random_state=0)
X3_train, X3_test, y3_train, y3_test = train_test_split(X3, y3, test_size = 0.2, random_state=0)
```
